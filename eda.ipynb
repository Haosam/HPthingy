```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.__version__)
```

```python
data = pd.read_csv('https://aisgaiap.blob.core.windows.net/aiap4-assessment/real_estate.csv')
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df.describe())
```

```python
# Having a Hypothesis - Earlier the sell date, the more expensive
sns.regplot(x=df["X1 transaction date"], y=df["Y house price of unit area"],line_kws={"color":"r"})
plt.xlim(2012, 2014)
plt.show()
```

```python
# Having a Hypothesis - Younger the house, the more expensive
sns.regplot(x=df["X2 house age"], y=df["Y house price of unit area"],line_kws={"color":"r"})
plt.show()
```

```python
# Having a Hypothesis - Nearer to MRT, the more expensive
sns.regplot(x=df["X3 distance to the nearest MRT station"], y=df["Y house price of unit area"],line_kws={"color":"r"})
plt.show()
```

```python
# Having a Hypothesis - More Convenience Stores, the more expensive
sns.regplot(x=df["X4 number of convenience stores"], y=df["Y house price of unit area"],line_kws={"color":"r"})
plt.show()
```

```python
sns.scatterplot(df["X6 longitude"], df['X5 latitude'], df["X1 transaction date"])
plt.show()
```

```python
sns.scatterplot(df["X6 longitude"], df['X5 latitude'], df["X2 house age"])
plt.show()
# Most of the older houses are located at Xindian District
```

```python
sns.scatterplot(df["X6 longitude"], df['X5 latitude'], df["X3 distance to the nearest MRT station"])
plt.show()
# The houses furtherest way from the MRT corresponds with the being far from Qizhang MRT
```

```python
sns.scatterplot(df["X6 longitude"], df['X5 latitude'], df["X4 number of convenience stores"])
plt.show()
# Convenience stores are densely located near the MRT
```

```python
sns.scatterplot(df["X6 longitude"], df['X5 latitude'], df["Y house price of unit area"])
plt.show()
# Houses in the densely located region tend to be more expensive
```
